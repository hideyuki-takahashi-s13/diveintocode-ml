# 【問題1】100日目の米粒の数

100日目には合計で米粒をいくつもらっているのかを計算するコードを作成し、答えを出力してください。また、1日目から100日目までの「その日にもらう米の数」と「その日までに累計で貰う米の数」の推移を折れ線グラフで表現してください。（縦軸が米の数、横軸が日数）

```python
'''
100日分の合計米粒
'''
    
N_RICE_1st_DAY = 1

TOTAL_DAYS = 100

list_n_rice_all_days = [N_RICE_1st_DAY]

for i in range(1, TOTAL_DAYS):    # 2日目以降の99日分を繰り返し処理
    list_n_rice_all_days.append(list_n_rice_all_days[-1] * 2)

print('100日分の合計米粒は、{}粒です。'.format(sum(list_n_rice_all_days)))
```

```python
'''
1日目から100日目までの「その日にもらう米の数」の折れ線グラフ
'''

import matplotlib.pyplot as plt
%matplotlib inline

# リストのインデックスは0から始まる為１〜１００のx軸を作成
number_of_days = list(range(1, 101))  

plt.title("n_rice_1st-100th_days")
plt.xlabel("number of days")
plt.ylabel("n_rice[n]")
plt.plot(number_of_days, list_n_rice_all_days)
plt.show()
```

```python
'''
「その日までに累計で貰う米の数」の折れ線グラフ
'''

list_n_rice_total_days = [N_RICE_1st_DAY]

n_rice_all_days = N_RICE_1st_DAY
for i in range(1, TOTAL_DAYS):    # 2日目以降の99日分を繰り返し処理
    n_rice_all_days *= 2
    list_n_rice_total_days.append(list_n_rice_total_days[-1] + n_rice_all_days)

# print(len(list_n_rice_total_days))

plt.title("n_rice_total_1st-100th_days")
plt.xlabel("number of days")
plt.ylabel("n_rice_total[n]")
plt.plot(number_of_days, list_n_rice_total_days)
plt.show()
```

# 【問題2】100日目以外の米粒の数

この逸話は30日だったり81日だったりと日数が曖昧です。どの話にも対応できるように日数を引数とした 関数 を作成しましょう。関数を作成した上で、コースを受講する日数の間でいくつ米粒をもらえるかを答えてください。前の問題同様に折れ線グラフも出力してください。

```python
def compute_sorori_shinzaemon(day=100):
    """
    曽呂利新左衛門の米の逸話を検証する関数。日にちごとの米の粒の数を計算する。

    Parameteres
    --------------
    day: int
        米を貰う日数 (default : 100)

    Returns
    --------------
    list_n_grains : list
        ある日に貰う米の数のリスト（リストの長さはdayに等しくなる）
    list_total_grains : list
        ある日までに累計で貰う米の数のリスト（リストの長さはdayに等しくなる）
    """
    # ここにコードを書く。passは何もしないことを意味する命令である。
    
    '''
     list_n_grainsを定義  
    '''  
    N_RICE_1st_DAY = 1

    TOTAL_DAYS = day

    list_n_rice_all_days = [N_RICE_1st_DAY]

    for i in range(1, TOTAL_DAYS):    # 2日目以降の99日分を繰り返し処理
        list_n_rice_all_days.append(list_n_rice_all_days[-1] * 2)
     
    list_n_grains = list_n_rice_all_days
    
    
    '''
　　　　 list_total_grainsを定義
　　　　'''
    list_n_rice_total_days = [N_RICE_1st_DAY]
    
    n_rice_all_days = N_RICE_1st_DAY
    for i in range(1, TOTAL_DAYS):    # 2日目以降の99日分を繰り返し処理
        n_rice_all_days *= 2
        list_n_rice_total_days.append(list_n_rice_total_days[-1] + n_rice_all_days)

    list_total_grains = list_n_rice_total_days
    
    return list_n_grains, list_total_grains

#コース期間４ヶ月（１ヶ月約３０日　×　４　　　＝　　　１２０日）にて引数を設定
course_days = 120
list_n_grains, list_total_grains = compute_sorori_shinzaemon(course_days)

print('コース受講日数での合計米粒は、{}粒です。'.format(list_total_grains[-1]))

# リストのインデックスは0から始まる為１〜１2０のx軸を作成
number_of_days = list(range(1, 121)) 
# list_total_grainsの折れ線グラフを出力
plt.title("total_grains_course_days")
plt.xlabel("number of days")
plt.ylabel("n_grains_total[n]")
plt.plot(number_of_days, list_total_grains)
plt.show()
```
