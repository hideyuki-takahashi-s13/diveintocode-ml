# 【問題2】ある地点の勾配を計算する関数を作成

「二次元配列と勾配問題」では各地点での勾配を最初に求めました。しかし、暗闇の富士山を想像すると、知ることができるのは今いる地点の勾配だけです。


「今いる地点の番号」と「今いる隣の地点の番号（今いる地点の番号-1）」の関係による勾配を求める関数を作成してください。


勾配は以下の式で求まりましたが、このデータでは「標高」がy、「その地点の番号」がxに相当します。

```python
def point_gradient(position):
    if position == 0:
        return 0
    else:
        d_p = position - (position - 1)    #　d_pは常に１だが、考え方（式）を記入
        d_e = fuji[position, 3] - fuji[position - 1, 3]
        return d_e / d_p

now_point = 136    #　「今いる地点の番号」　を変数に代入
point_gradient(now_point)    #　「今いる地点の番号」を引数に取り関数を実行
```

# 【問題3】移動先の地点を計算する関数を作成 
今いる地点の勾配の情報を元に、次に移動先の地点を計算する関数を作成してください。<br>
数式で表すと 「移動先の地点 = 今いる地点 - $\alpha$ ×今いる地点の勾配」 となります。
<br>
「今いる地点の勾配」は問題2で作成した関数の返り値を使用してください。<br>
$\alpha$ は0より大きい任意の値をとるハイパーパラメータです。<br>
機械学習の分野では学習率という名前で呼ばれます。この値によって下り方が変わります。<br>
ここでは例えば0.2に設定してください。

```python
from decimal import Decimal, ROUND_HALF_UP

h_p = 0.2    #　ハイパーパラメータ
now_point = 136    #　「今いる地点の番号」　を変数に代入
point_gradient(now_point)    #　「今いる地点の番号」を引数に取り、問題２の関数を実行（勾配を返す）

def move_point(now_point, h_p):
    move_point = now_point - h_p * point_gradient(now_point)
    
    # Decimalを使用し四捨五入
    move_point_Deci = Decimal(str(move_point)).quantize(Decimal('0'), rounding=ROUND_HALF_UP)
   
    # if文にてマイナスや300以上の数値にならないよう処理
    if move_point_Deci < 0:
        return 0
    elif move_point_Deci> 299:
        return 299
    else:
        return int(move_point_Deci)

move_point(now_point, h_p)
```

# 【問題4】山を下る関数を作成
山を下る関数を作成してください。最初の地点を受け取り、問題3の関数によって地点を更新していきます。移動毎の地点はリストなどに記録して、これを関数の返り値としてください。


最初の「今いる地点」は富士山の頂上付近、136番の地点（標高3434.59[m]）とします。ここから山を下っていきましょう。


《実装上の注意》


問題3で四捨五入を行っているため、移動先の地点が今いる地点と同じ場合になることがあります。移動量が減ってきたということなので、その時点で山を下りきったと判断してください。実際の勾配降下法ではこのようなことは起きません。

```python
first_point = 136

def down_mountain(first_point, h_p):
    point_list = [first_point]    #移動毎の地点をリストに記録
    while True:
        next_point = move_point(point_list[-1], h_p)    #問題３の関数で次の地点を返す
        point_list.append(next_point)    #次の地点をリストに格納
        if next_point in point_list[:-1]:    #リストを更新していき、同じ地点になった時点で終了
            return point_list    #リストを返す
            break

down_mountain(first_point, h_p)
```

# 【問題5】下山過程の可視化

問題4における下山過程をMatplotlibを使いグラフで可視化してください。<br>
《可視化例》<br>
- 断面図の中に移動した地点を点で表示する<br>
- 繰り返し毎の標高や勾配を折れ線グラフにする

```python
def visualization(first_point):

    #移動した地点の標高リストを作成
    move_elevation_list = []
    for i in down_mountain(first_point, h_p):
        move_elevation_list.append(fuji[i, 3])

    #断面図の中に移動した地点を点で表示する
    plt.xlabel("position")
    plt.ylabel("elevation[m]")
    plt.title("Mt. Fuji & move_point")
    plt.plot(fuji[:, 0], fuji[:, 3], label="Mt. Fuji")    #富士山断面図のプロット
    plt.plot(down_mountain(first_point, h_p), move_elevation_list, "bo", label="move_point", color='r')    #移動した地点のプロット
    plt.legend()
    plt.show()

    #移動した地点の勾配リストを作成
    move_gradient_list = []
    for i in down_mountain(first_point, h_p):
        move_gradient_list.append(int(point_gradient(i)))

    #繰り返し毎の標高や勾配を折れ線グラフにする
    fig, ax1 = plt.subplots()
    ax2 = ax1.twinx()
    plt.title("move_elevation & move_gradient")
    ax1.set_xlabel("position")
    ax2.set_xlabel("position")
    ax1.set_ylabel("elevation[m]")
    ax2.set_ylabel("gradient")
    ax1.plot(down_mountain(first_point, h_p), move_elevation_list, label="move_elevation", color='r')    #移動した地点のプロット
    ax2.plot(down_mountain(first_point, h_p), move_gradient_list, label="move_gradient")    #移動した地点の勾配のプロット
    ax1.legend(bbox_to_anchor=(0, 1), loc='upper left', borderaxespad=0.5, fontsize=10)
    ax2.legend(bbox_to_anchor=(0, 0.9), loc='upper left', borderaxespad=0.5, fontsize=10)
    plt.show()
    
visualization(first_point)
```

# 【問題6】初期値の変更
問題4では最初の「今いる地点」を136としましたが、これを他の全ての点で計算してください。<br>
そして、いくつかの初期値に対してどのような下山過程となるか、問題5のように可視化してください。<br>
機械学習においては 初期値 をどう設定するかが結果に影響を与えることが確認できます。<br>
《初期値とは》<br>
今回の例での最初の地点のような値を初期値と呼びます。これを変更すると結果が変化します。<br>
例えば136と142では富士山の異なるの斜面を下ることになります

```python
#全ての地点で計算

for i in fuji[:,0]:
    first_point = int(i)
    def down_mountain(first_point, h_p):
        point_list = [first_point]    #移動毎の地点をリストに記録
        while True:
            next_point = move_point(point_list[-1], h_p)    #問題３の関数で次の地点を返す
            point_list.append(next_point)    #次の地点をリストに格納
            if next_point in point_list[:-1]:    #リストを更新していき、同じ地点になった時点で終了
                return point_list    #リストを返す
                break

    print(down_mountain(first_point, h_p))
```

```python
#いくつかの初期値の可視化

first_point = 50    #初期値

visualization(first_point)
```

```python
#いくつかの初期値の可視化

first_point = 150    #初期値

visualization(first_point)
```

```python
#いくつかの初期値の可視化

first_point = 250    #初期値

visualization(first_point)
```

# 【問題7】（アドバンス問題）ハイパーパラメータの変更
ハイパーパラメータ $\alpha$ を変化させるとどのような下山過程となるか、問題5のように可視化してください。<br>
そして、ハイパーパラメータにより結果が大きく異なることを確認してください。

```python
h_p = 0.5    #　ハイパーパラメータ
first_point = 136    #初期値

move_point(first_point, h_p)
down_mountain(first_point, h_p)
visualization(first_point)
```

```python
#比較用

h_p = 0.2    #　ハイパーパラメータ
first_point = 136    #初期値

move_point(first_point, h_p)
down_mountain(first_point, h_p)
visualization(first_point)
```
