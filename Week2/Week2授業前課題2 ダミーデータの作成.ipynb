# 【問題1】乱数の作成
平均が(-3, 0)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。

```python
# 乱数500個を作成

import numpy as np

mean = np.array([-3, 0])
cov = np.array([[1.0, 0.8], [0.8, 1.0]])

np.random.seed(0)
random_ndarray = np.random.multivariate_normal(mean, cov, 500)

random_ndarray.shape
```

# 【問題2】散布図による可視化
問題1で作成したデータ点を散布図により可視化してください。

```python
# 散布図により可視化

import matplotlib.pyplot as plt
%matplotlib inline

x1 = random_ndarray[:, 0]
x2 = random_ndarray[:, 1]

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("scatter")
plt.scatter(x1, x2)
plt.show()
```

# 【問題3】ヒストグラムによる可視化
問題1で作成したデータをヒストグラムにより可視化してください。<br>
各次元ごとに作成し、2つのヒストグラムとしてください。<br>
それぞれのヒストグラムの横軸の範囲はplt.xlim()を用いて揃えるようにしてください。

```python
plt.xlabel("x1")
plt.ylabel("frequency")
plt.title("histgram of x1")
plt.xlim(-6, 3)
plt.hist(x1)
plt.show()
```

```python
plt.xlabel("x2")
plt.ylabel("frequency")
plt.title("histgram of x2")
plt.xlim(-6, 3)
plt.hist(x2)
plt.show()
```

# 【問題4】データの追加
新たに平均が(0, -3)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。

```python
mean = np.array([0, -3])
cov = np.array([[1.0, 0.8], [0.8, 1.0]])

np.random.seed(0)
random_ndarray_1 = np.random.multivariate_normal(mean, cov, 500)

random_ndarray_1.shape
```

そして、問題1、4それぞれのデータをひとつの散布図として可視化してください。凡例として問題1のものは0、問題2のものは1を表示してください。

```python
# 散布図により可視化

import matplotlib.pyplot as plt
%matplotlib inline

# 問題１のデータ
x1_0 = random_ndarray[:, 0]
x2_0 = random_ndarray[:, 1]

# 問題４のデータ
x1_1 = random_ndarray_1[:, 0]
x2_1 = random_ndarray_1[:, 1]

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("scatter")
plt.scatter(x1_0, x2_0, label="0")
plt.scatter(x1_1, x2_1, label="1")
plt.legend()
```

# 【問題5】データの結合
データはまとめておいた方が後々扱いやすいです。問題1、4で作成したndarrayを 結合 し、(1000, 2)のndarrayとしてください。<br>
結合はnp.concatenate()やnp.vstack()を使うことで行えます。

```python
# np.concatenate()を使用
print(np.concatenate((random_ndarray, random_ndarray_1), axis=0).shape)
np.concatenate((random_ndarray, random_ndarray_1), axis=0)
```

```python
# np.vstack()を使用
print(np.vstack((random_ndarray, random_ndarray_1)).shape)
np.vstack((random_ndarray, random_ndarray_1))
```

# 【問題6】ラベル付け
ひとまとめになった1000個のデータそれぞれに対して、問題1、4どちらで作成したものなのかを示す ラベル 付けを行ってください。<br>
問題1のものには0、問題4のものには1を対応させます。<br>
0と1を含むラベルの列を新たに追加し、(1000, 3)のndarrayを作成してください。
機械学習に使用するデータセットはこのような形になっていることが多いです。

```python
random_ndarray_label = np.insert(random_ndarray, 2, 0, axis=1)
random_ndarray_1_label = np.insert(random_ndarray_1, 2, 1, axis=1)

print(np.concatenate((random_ndarray_label, random_ndarray_1_label), axis=0).shape)
np.concatenate((random_ndarray_label, random_ndarray_1_label), axis=0)
```
