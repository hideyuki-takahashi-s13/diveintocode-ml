# 【問題1】データの取得
以下のドキュメントを参考にしながら、irisデータセット読み込みましょう。<br>
データが読み込めたら、説明変数を下のようなPandasのDataFrame型変数Xに格納してください。<br>列の名前も以下を参考にしてください。<br>
目的変数も同様にyに格納してください。

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = pd.DataFrame(iris.target, columns=["Species"])

X, y
```

# 【問題6】図の作成
データは可視化して確認することが大切です。次の3つを行ってください。<br>作図のライブラリはMatplotlibの他seabornもよく使われます。
- ラベルごとのサンプル数を円グラフにする（パーセンテージまで表示する）
- 特徴量を一つずつ選び、箱ひげ図を使ってラベルごとのデータの分布を可視化する
- 特徴量を一つずつ選び、バイオリン図を使ってラベルごとのデータの分布を可視化する

箱ひげ図とバイオリン図は似た図を表示します。これらはどう違い、それぞれどのような利点があるのかについて説明してください。

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# ラベルごとのサンプル数を円グラフにする（パーセンテージまで表示する）
list_columns = list(df.columns.values)    #カラム名を抽出してリストへ格納
n_samples = [len(df.loc[:, i]) for i in list_columns]    #各ラベル毎のサンプル数を抽出してリストへ格納

fig, ax = plt.subplots()

ax.pie(n_samples[0:-1], labels=list_columns[0:-1], startangle=90, counterclock=False,
              shadow=True, autopct="%1.2f%%", explode=[0, 0, 0, 0.2])

plt.show()
```

```python
# 特徴量を一つずつ選び、箱ひげ図を使ってラベルごとのデータの分布を可視化する
list_data = [df.loc[:, i].astype(float) for i in list_columns]    #各カラム毎の値を抽出してfloatへ変換（データフレームでは描画できなかった為）しリストへ格納

fig, ax = plt.subplots(figsize=(10, 5), facecolor='w')

ax.boxplot(list_data[0:-1], labels=list_columns[0:-1])
# df.iloc[:, 0:-1].boxplot()    #データフレーム型を使う方法

plt.show()
```

```python
# 特徴量を一つずつ選び、バイオリン図を使ってラベルごとのデータの分布を可視化する

fig, ax = plt.subplots(figsize=(10, 5))

ax.violinplot(list_data[0:-1])
ax.set_xticks([1, 2, 3, 4])
ax.set_xticklabels(list_columns[0:-1])

# sns.violinplot(data=df.iloc[:, 0:-1])    #データフレーム型を使う方法

plt.show()
```

### 箱ひげ図には平均/中央値と四分位範囲という要約統計量のみが含まれるのに対し、バイオリン図にはすべての分布データが含まれる。その為、箱ひげ図はデータがどのあたりに集中しているかを捉え易いのに対し、バイオリン図は全体の分布を捉え易い。

# 【問題7】特徴量間の関係の確認
特徴量間の関係を見ていきます。次の4つを行ってください。<br>
- 縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る
- 全ての組み合わせの散布図を一度に表示させる、散布図行列を作る
- ４つの特徴量に関する相関係数行列を作る
- 相関係数行列のヒートマップを作る

相関係数行列はDataFrameのcorr()メソッドを使うことで求められます。

```python
# 縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る
fig = plt.subplots(figsize=(10, 10))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='Species', data=df)
plt.show()
```

```python
# 全ての組み合わせの散布図を一度に表示させる、散布図行列を作る
sns.pairplot(df, hue='Species', vars=df.iloc[:, 0:-1])
plt.show()
```

```python
# ４つの特徴量に関する相関係数行列を作る
corr_df = df.iloc[:, 0:-1].corr(method='pearson')
corr_df
```

```python
# 相関係数行列のヒートマップを作る
fig, ax = plt.subplots(figsize=(10,8))
g = sns.heatmap(corr_df, vmin=-1.0, vmax=1.0, center=0, annot=True, fmt='.3f', cmap="RdBu_r")
ax.set_ylim(4, 0)    #グラフの上下が切れたので追加
g.set_yticklabels(g.get_yticklabels(), rotation =0)    #y軸ラベルが見易くなるよう回転
plt.show()
```
