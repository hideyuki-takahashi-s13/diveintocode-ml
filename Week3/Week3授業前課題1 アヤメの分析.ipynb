# 【問題1】データの取得
以下のドキュメントを参考にしながら、irisデータセット読み込みましょう。<br>

目的変数も同様にyに格納してください。

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = pd.DataFrame(iris.target, columns=["Species"])

X, y
```

#### ・特徴量を一つずつ選び、バイオリン図を使ってラベルごとのデータの分布を可視化する

```python
# 特徴量を一つずつ選び、バイオリン図を使ってラベルごとのデータの分布を可視化する

"""
fig, ax = plt.subplots(2, 2, figsize=(20, 10))
plt.xlabel("Species")
plt.ylabel("length (cm)")
plt.title("sepal length")
df_sepal_length = df.pivot(columns='Species', values='sepal length (cm)')
sns.violinplot(data=df_sepal_length)
plt.show()
"""

columns_list = [df.columns.values[2], df.columns.values[3], df.columns.values[0], df.columns.values[1]] #箱ひげ図に順番を合わせる為に並べ替え
df_concat_axis0 = pd.DataFrame() #空のデータフレームを用意
for i in columns_list: #sns.catplot用に、このfor文の中で新しいデータフレーム構造を作成
    df_pivot = df.pivot(columns='Species', values=i)
    df_columns = pd.DataFrame([i]*150)
    df_concat_axis1 = pd.concat([df_pivot, df_columns], axis=1)
    df_concat_axis0 = pd.concat([df_concat_axis0, df_concat_axis1])

df_concat_axis0.columns = [0, 1, 2, "feature"] #カラムを付与

violin = sns.catplot(data=df_concat_axis0, kind="violin", col="feature", col_wrap=2)
(violin.set_axis_labels("Species", "length & width(cm)").set_titles("{col_name}"))
```

### 箱ひげ図には平均/中央値と四分位範囲という要約統計量のみが含まれるのに対し、バイオリン図にはすべての分布データが含まれる。その為、箱ひげ図はデータがどのあたりに集中しているかを捉え易いのに対し、バイオリン図は全体の分布を捉え易い。

# 【問題7】特徴量間の関係の確認
特徴量間の関係を見ていきます。次の4つを行ってください。<br>
- 縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る
- 全ての組み合わせの散布図を一度に表示させる、散布図行列を作る
- ４つの特徴量に関する相関係数行列を作る
- 相関係数行列のヒートマップを作る

相関係数行列はDataFrameのcorr()メソッドを使うことで求められます。

```python
# 縦軸と横軸に特徴量を１つずつ取り、種類毎に色分けした散布図を作る
fig = plt.subplots(figsize=(10, 10))
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='Species', data=df)
plt.show()
```

```python
# 全ての組み合わせの散布図を一度に表示させる、散布図行列を作る
sns.pairplot(df, hue='Species', vars=df.iloc[:, 0:-1])
plt.show()
```

```python
# ４つの特徴量に関する相関係数行列を作る
corr_df = df.iloc[:, 0:-1].corr(method='pearson')
corr_df
```

```python
# 相関係数行列のヒートマップを作る
fig, ax = plt.subplots(figsize=(10,8))
g = sns.heatmap(corr_df, vmin=-1.0, vmax=1.0, center=0, annot=True, fmt='.3f', cmap="RdBu_r")
# ax.set_ylim(4, 0)    #グラフの上下が切れたので追加
g.set_yticklabels(g.get_yticklabels(), rotation =0)    #y軸ラベルが見易くなるよう回転
plt.show()
```
