## 前処理

```python
from sklearn.model_selection import train_test_split
```

```python
#ベースラインモデル用に、説明変数２つと目的変数を抜き出す。
X = df[["DAYS_BIRTH","AMT_ANNUITY"]].fillna(df['AMT_ANNUITY'].mean()) #AMT_ANNUITYの欠損値を平均で補完
y = df.loc[:, 'TARGET']

# 訓練データと検証データの分割。訓練データ75%、検証データ25%として分割する。
X = np.array(X)
y = np.array(y)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=123)
display(X_train.shape)
display(X_test.shape)
display(y_train.shape)
display(y_test.shape)
```

```python
# 分割されている割合の確認
u, counts = np.unique(y_train, return_counts=True)
print(u)
print(counts)
```

```python
#標準化
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(X_train) #訓練用のデータでfit
X_train_std = scaler.transform(X_train) #訓練用データをtransform
X_train_std
```

```python
X_test_std = scaler.transform(X_test) #検証用データをtransform
X_test_std
```

## 学習

```python
#アルゴリズム　インポート
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier

#評価指標　インポート
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import auc
from sklearn import metrics
from sklearn.metrics import classification_report
```

```python
#ランダムフォレストを使用
clf_rfc = RandomForestClassifier(random_state=0)
clf_rfc.fit(X_train_std, y_train) # 学習
rfc_pred = clf_rfc.predict(X_test_std) #クラスの予測
rfc_pred_proba = clf_rfc.predict_proba(X_test_std) #クラスの予測確率
rfc_pred_proba_posi = rfc_pred_proba[:, 0] #クラスの予測確率（ポジティブクラスを抜粋）
rfc_pred_score = clf_rfc.score(X_test_std, y_test)#スコア（平均精度）
print(rfc_pred)
print(rfc_pred_proba)
print(rfc_pred_proba_posi)
print(rfc_pred_score)
```

```python
# 評価
print("正解率：{:.2f}".format(accuracy_score(y_test, rfc_pred)))
print("適合率：{:.2f}".format(precision_score(y_test, rfc_pred, pos_label=0)))
print("再現率：{:.2f}".format(recall_score(y_test, rfc_pred, pos_label=0)))
print("F値：{:.2f}".format(f1_score(y_test, rfc_pred, pos_label=0)))
print("混同行列：")
print(confusion_matrix(y_test, rfc_pred))
print("classification_report：") #評価をまとめて出力するやつ
print(classification_report(y_test, rfc_pred))
```

## 検証

```python
# ROC曲線の計算
fpr, tpr, thresholds = metrics.roc_curve(y_test, rfc_pred_proba_posi, pos_label=0)
print(fpr, tpr, thresholds)
```

```python
# ROC曲線をグラフへプロット
plt.plot(fpr, tpr, marker='o')
plt.xlabel('FPR: False positive rate')
plt.ylabel('TPR: True positive rate')
plt.grid()
```

```python
#AUC（Area Under the Curve）を計算
auc_value = auc(fpr, tpr)
print("AUC値は{:.2f}です。".format(auc_value))
```

### ベースモデルのAUC値は０.５４となった。

### オブジェクトデータを説明変数として使用する為、数値データへ変換する。

```python
#オブジェクトデータの欠損値有無を確認
df_object = df.select_dtypes(include=object)
df_object.info()
```

```python
#オブジェクトデータの要素のユニーク数を確認
df_object.nunique()
```

```python
#オブジェクトデータの要素のユニーク名を確認
for i in df_object.columns:
    print("{}：　{}".format(i, df_object[i].unique()))
```

```python
#欠損値がある説明変数を除去する
df_object_drop =df_object.dropna(axis=1)
df_object_drop
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
le = LabelEncoder()
```

```python
#欠損値を除去したオブジェクトデータをラベルエンコーディングする。
le_df_object_drop = np.array([])
for i in df_object_drop.columns:
    le.fit(df_object_drop[i])
    le_df_object_drop = np.append(le_df_object_drop, le.transform(df_object_drop[i]))

le_df_object_drop = pd.DataFrame(le_df_object_drop.reshape([10, -1]).T) #データフレーム化
le_df_object_drop.columns = df_object_drop.columns #カラム名を付与
le_df_object_drop
```

```python
#ラベルエンコーディングしたデータをOne-hotエンコーディングする。
ohe_df_object_drop = pd.DataFrame()
for i in df_object_drop.columns:
    ohe_df = pd.get_dummies(df.loc[:, i], prefix=i)
    ohe_df_object_drop = pd.concat([ohe_df_object_drop, ohe_df], axis=1)

ohe_df_object_drop
```

```python
#学習から検証までを関数化
def learning_to_verification(X, y):
    # 訓練データと検証データの分割。訓練データ75%、検証データ25%として分割する。
    X = np.array(X)
    y = np.array(y)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=123)
    
    #標準化
    scaler.fit(X_train) #訓練用のデータでfit
    X_train_std = scaler.transform(X_train) #訓練用データをtransform
    X_test_std = scaler.transform(X_test) #検証用データをtransform

    #ランダムフォレストで学習〜予測
    clf_rfc = RandomForestClassifier(n_estimators=100, random_state=0)
    clf_rfc.fit(X_train_std, y_train) # 学習
    rfc_pred = clf_rfc.predict(X_test_std) #クラスの予測
    rfc_pred_proba = clf_rfc.predict_proba(X_test_std) #クラスの予測確率
    rfc_pred_proba_posi = rfc_pred_proba[:, 0] #クラスの予測確率（ポジティブクラスを抜粋）
    rfc_pred_score = clf_rfc.score(X_test_std, y_test)#スコア（平均精度）

    # 評価
    print("正解率：{:.2f}".format(accuracy_score(y_test, rfc_pred)))
    print("適合率：{:.2f}".format(precision_score(y_test, rfc_pred, pos_label=0)))
    print("再現率：{:.2f}".format(recall_score(y_test, rfc_pred, pos_label=0)))
    print("F値：{:.2f}".format(f1_score(y_test, rfc_pred, pos_label=0)))
    print("混同行列：")
    print(confusion_matrix(y_test, rfc_pred))
    print("classification_report：") #評価をまとめて出力するやつ
    print(classification_report(y_test, rfc_pred))

    # ROC曲線の計算
    fpr, tpr, thresholds = metrics.roc_curve(y_test, rfc_pred_proba_posi, pos_label=0)

    # ROC曲線をグラフへプロット
    plt.plot(fpr, tpr, marker='o')
    plt.xlabel('FPR: False positive rate')
    plt.ylabel('TPR: True positive rate')
    plt.grid()

    #AUC（Area Under the Curve）を計算
    auc_value = auc(fpr, tpr)
    print("AUC値は{:.2f}です。".format(auc_value))
```

```python
#ベースラインモデル
X = df[["DAYS_BIRTH","AMT_ANNUITY"]].fillna(df['AMT_ANNUITY'].mean())
y = df.loc[:, 'TARGET']

learning_to_verification(X, y) #学習から検証までを関数で実行
```

### No.1:　ベースラインモデルに、One-hotエンコーディングしたオブジェクトの説明変数(欠損値が無い変数全て)を加える。

```python
X =  pd.concat([df[["DAYS_BIRTH","AMT_ANNUITY"]].fillna(df['AMT_ANNUITY'].mean()), ohe_df_object_drop], axis=1)
y = df.loc[:, 'TARGET']

learning_to_verification(X, y) #学習から検証までを関数で実行
```

### ベースラインモデルに、One-hotエンコーディングしたオブジェクトの説明変数(欠損値が無い変数全て)を加えたところ、AUC値は0.54から0.57へ若干向上した。

### No.2:　欠損値が無い説明変数全てで学習。

```python
df_num = df.select_dtypes(include="number") #数値データを抽出
df_num_dropna = df_num.dropna(axis=1) #欠損値があるデータ列（カラム）を除去
df_num_drop = df_num_dropna.drop(['TARGET', "SK_ID_CURR"], axis=1) #目的変数、IDを除去
X =  pd.concat([df_num_drop, ohe_df_object_drop], axis=1) #数値データとOne-hotエンコーディングしたオブジェクトデータを結合
y = df.loc[:, 'TARGET']

learning_to_verification(X, y) #学習から検証までを関数で実行
```

### 欠損値が無い説明変数全てで学習を実施したところ、ベースラインモデルと比較し、AUC値は0.54から0.64へ向上した。

### No.3:　No.2に対し欠損値が有る数値データの説明変数を全て加える。

```python
df_num = df.select_dtypes(include="number") #数値データを抽出
df_num_fillmedian = df_num.fillna(df_num.median()) #欠損値を中央値で補完
df_num_drop = df_num_fillmedian.drop(['TARGET', "SK_ID_CURR"], axis=1) #目的変数、IDを除去

X =  pd.concat([df_num_drop, ohe_df_object_drop], axis=1) #数値データとOne-hotエンコーディングしたオブジェクトデータを結合
y = df.loc[:, 'TARGET']

learning_to_verification(X, y) #学習から検証までを関数で実行
```

### 数値データの欠損値を中央値で補完し、全ての数値データと欠損値が無いオブジェクトデータで学習を実施したところ、No.2と比較し、AUC値は0.64から0.71へ向上した。

### No.4:　相関が強い説明変数(数値データ)を減らす。

```python
#相関行列ヒートマップで相関が高い組み合わせの有無を俯瞰する。
corr_num = df_num_fillmedian.corr(method='pearson')

fig, ax = plt.subplots(figsize=(200,200))
g = sns.heatmap(corr_num, vmin=-1.0, vmax=1.0, center=0, annot=True, fmt='.2f', cmap="RdBu_r")

plt.show()
```

```python
#相関が高い（相関係数：0.8以上）組み合わせを抽出する。
corr_high = np.where(abs(corr_num) >= 0.8)
corr_high_indices = [(corr_num.index[x], corr_num.columns[y]) for x, y in zip(*corr_high) if x != y and x < y]
corr_high_indices = pd.DataFrame(corr_high_indices)
corr_high_unique = corr_high_indices[1].unique()
corr_high_unique
```

```python
df_num_drop = df_num_fillmedian.drop(['TARGET', "SK_ID_CURR"], axis=1) #目的変数、IDを除去
df_num_high_drop =  df_num_drop.drop(corr_high_unique, axis=1) #説明変数同士で相関が高い変数（の半数）を除去

X =  pd.concat([df_num_high_drop, ohe_df_object_drop], axis=1) #数値データとOne-hotエンコーディングしたオブジェクトデータを結合
y = df.loc[:, 'TARGET']

learning_to_verification(X, y) #学習から検証までを関数で実行
```

### No.3より、相関が高い説明変数(相関係数：0.8以上)を抽出し半数を削除したが、AUC値は0.71で変わらず。改善はしなかったが悪くもならなかった。

### No.5:　No.4より重要度が低い説明変数(数値データ)を減らす。

```python
#重要度を出す

X =  pd.concat([df_num_high_drop, ohe_df_object_drop], axis=1) #数値データとOne-hotエンコーディングしたオブジェクトデータを結合
y = df.loc[:, 'TARGET']

# 訓練データと検証データの分割。訓練データ75%、検証データ25%として分割する。
X = np.array(X)
y = np.array(y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=123)
    
#標準化
scaler.fit(X_train) #訓練用のデータでfit
X_train_std = scaler.transform(X_train) #訓練用データをtransform
X_test_std = scaler.transform(X_test) #検証用データをtransform

#ランダムフォレストで学習〜予測
clf_rfc = RandomForestClassifier(n_estimators=100, random_state=0)
clf_rfc.fit(X_train_std, y_train) # 学習

fti = clf_rfc.feature_importances_ #重要度
print(fti)
```

### No.4より、重要度が高い説明変数(0.01以上)を抽出し学習したが、今回もAUC値は0.71で変わらず。No.4及び5にて説明変数を大きく削除したが、AUC値が変わらなかったことから、やはり一部の説明変数の寄与度が大きい。

### テストデータに対しても予測を行う。
